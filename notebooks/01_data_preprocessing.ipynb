```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported!")
```

```python
# Download the Cleveland Heart Disease dataset directly from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# Column names (the file has no headers, so we add them)
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs',
                'restecg', 'thalach', 'exang', 'oldpeak', 'slope',
                'ca', 'thal', 'target']

df = pd.read_csv(url, header=None, names=column_names)

print("✅ Dataset loaded!")
print("Shape:", df.shape)
df.head()
```

```python
# The dataset uses '?' for missing values
df.replace('?', np.nan, inplace=True)

# Convert everything to numbers
df = df.apply(pd.to_numeric, errors='coerce')

# Check missing values
print("Missing values per column:")
print(df.isnull().sum())

# Fill missing with median
df.fillna(df.median(), inplace=True)
print("\n✅ Missing values fixed!")
```

```python
# X = inputs, y = output (heart disease yes/no)
X = df.drop('target', axis=1)
y = df['target']

# Convert to binary: 0 = no disease, 1 = disease
y = y.apply(lambda x: 1 if x > 0 else 0)

print("Feature columns:", list(X.columns))
print("\nTarget distribution:")
print(y.value_counts())
print("\n0 = No Heart Disease, 1 = Heart Disease")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("✅ Done!")
print(f"Training samples : {X_train.shape[0]}")
print(f"Testing samples  : {X_test.shape[0]}")
```

```python
import os
os.makedirs('../models', exist_ok=True)

joblib.dump(scaler, '../models/scaler.pkl')
print("✅ Scaler saved!")
```
